```python
from scipy.special import erf
import numpy as np

import matplotlib.pyplot as pl
import matplotlib as mpl
%matplotlib inline
```

```python
def probit(x):
    return 0.5 * (1.0 + erf(x / np.sqrt(2.0))) * (1 - 2e-3) + 1e-3
```

```python
x = np.arange(-5,5,0.01)
y = probit(x)
pl.figure(figsize=(15,5))
pl.plot(x,y); 
pl.xlabel('x')
pl.ylabel('probit(x)')
pl.title('invlink')
pl.show()
```
